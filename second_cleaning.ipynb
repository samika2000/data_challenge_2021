```python
from scipy.stats import mode
outlet_typemode=data.pivot_table(values='Outlet_Type', columns='Outlet_Identifier', aggfunc=lambda x:x.mode())
outlet_typemode
```

```python
def input_outlet_type(cols):
    outlet_type=cols[0]
    identifier=cols[1]
    if(pd.isnull(outlet_type)):
        return outlet_typemode.loc['Outlet_Type'][outlet_typemode.columns==identifier].get(0)
    else:
        return outlet_type

# print('Og missing: %d'%sum(data['Outlet_Size'].isnull()))
data['Outlet_Type']=data[['Outlet_Type', 'Outlet_Identifier']].apply(input_outlet_type,axis=1)
# print('Final missing: %d'%sum(data['Outlet_Type'].isnull()))

```

```python
data.isnull().sum()/data.shape[0]*100
print(list(data['Outlet_Type'].unique()))
```

```python
item_type=data.pivot_table(values='Item_Type', columns='Item_Identifier', aggfunc=lambda x:x.mode())
# outlet_sizemode
def fill_item_type(cols):
    it=cols[0]
    idvals=cols[1]
    if(pd.isnull(it)):
        return item_type.loc['Item_Type'][item_type.columns==idvals].get(0)
    else:
        return it

data['Item_Type']=data[['Item_Type', 'Item_Identifier']].apply(fill_item_type,axis=1)
data.isnull().sum()/data.shape[0]*100
```

```python
print(list(data['Item_Type'].unique()))
```

```python

outlet_idmode=data.pivot_table(values='Outlet_Identifier', columns='Outlet_Establishment_Year', aggfunc=lambda x:x.mode())


def input_outlet_id(cols):
    outlet_id=cols[0]
    yr=cols[1]
    if(pd.isnull(outlet_id)):
        return outlet_idmode.loc['Outlet_Identifier'][outlet_idmode.columns==yr].get(0)
    else:
        return outlet_id

# print('Og missing: %d'%sum(data['Outlet_Size'].isnull()))
data['Outlet_Identifier']=data[['Outlet_Identifier', 'Outlet_Establishment_Year']].apply(input_outlet_id,axis=1)
```
